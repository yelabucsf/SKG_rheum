```python
#Load 10x data



#Update all these inputs
#START
no_wells = sample_list.shape[0]

RNA_files = np.array(sample_list['sample'])
TCR_files = np.array(sample_list['TCR'])
mouse = np.array(sample_list['mouse'])
genotype = np.array(sample_list['genotype'])
GFP = np.array(sample_list['GFP'])


RNA_path_prefix =  #Fill in path to cellranger output directory
RNA_path_suffix = "/outs/filtered_feature_bc_matrix"

TCR_path_prefix = #Fill in path to cellranger output directory
TCR_path_suffix = "/analysis_vdj/outs/filtered_contig_annotations.csv"
#END

#Initialize list to store adata objects
RNA_adata = [None] * no_wells
TCR_csv = [None] * no_wells
TCR_cells_csv = [None] * no_wells
barcode_counts = [None] * no_wells

#Initialize dataframe to store cell counts
#height = 6
height = 4
width = no_wells
cell_count_df = pd.DataFrame(0, index=range(height), columns=range(width))
cell_count_df.rename(index={0:'Cells',1:'TCR cells', 2:'% TCR coverage', 3:'% with 1 TRA/TRB'}, inplace=True)
#Rename the columns
cell_count_df.columns = np.array(sample_list['sample_names'])


for i in range(no_wells):
    start = time.time()
    
    well = RNA_files[i]
    print(well)
    
    #Load RNA adata
    path = RNA_path_prefix + RNA_files[i] + RNA_path_suffix
    RNA_adata[i] = sc.read_10x_mtx(path, var_names='gene_symbols', cache=True)
    print("Loaded RNA")
    
    ##Load TCR
    path = TCR_path_prefix + TCR_files[i] + TCR_path_suffix
    TCR_csv[i] = pd.read_csv(path)
    
    #TCR Outputs
    #Barcodes with TCR
    unique_barcodes = np.unique(TCR_csv[i]["barcode"])
    print("Unique barcodes with TCR seq", len(unique_barcodes))
    
    #Cells with TCR
    TCR_csv_cells = TCR_csv[i].loc[TCR_csv[i]['barcode'].isin(RNA_adata[i].obs.index)]
    print("Cells: ", RNA_adata[i].shape[0])
    print("Cells with TCR: ", len(np.unique(TCR_csv_cells['barcode'])))
    
    #CELLS WITH CDR3 AND JUST 1 TRA/TRB
    TCR_csv_cells = TCR_csv_cells.loc[TCR_csv_cells['cdr3'] != "None"]
    unique_barcodes_cells = np.unique(TCR_csv_cells["barcode"])
    barcode_counts_all = {}
    for barcode in unique_barcodes_cells:
        TRA = np.sum(TCR_csv_cells[(TCR_csv_cells["barcode"] == barcode)]["chain"] == "TRA")
        TRB = np.sum(TCR_csv_cells[(TCR_csv_cells["barcode"] == barcode)]["chain"] == "TRB")
        Multi = np.sum(TCR_csv_cells[(TCR_csv_cells["barcode"] == barcode)]["chain"] == "Multi")
        barcode_counts_all[barcode] =[TRA,TRB,Multi]
    
    TCR_cells_csv[i] = TCR_csv_cells
    barcode_counts[i] = barcode_counts_all
    #frequency = np.unique(CD3_TCR_singlets['raw_clonotype_id'], return_counts = True)[1][1:]
    print("Cells with cdr3:", len(unique_barcodes_cells))
    tra_trb_singles = len([num for num in list(barcode_counts_all.values()) if num[0] <= 1 and num[1] <= 1])
    print("Cells with just 1 TRA/TRB:", tra_trb_singles)

    
    print("Added singlet info")
    cell_count_df[well] = [RNA_adata[i].shape[0],
                           len(unique_barcodes_cells),len(unique_barcodes_cells)/RNA_adata[i].shape[0], tra_trb_singles/RNA_adata[i].shape[0]]
    
    print("Add cell annotations")
    

    RNA_adata[i].obs['mouse'] = mouse[i] #fixed
    RNA_adata[i].obs['genotype'] = genotype[i] #fixed
    RNA_adata[i].obs['GFP'] = GFP[i] #fixed
    
    end = time.time()
    print(end-start, "sec")
```

```python
#Initialize dataframe to store cell counts
height = 4
width = no_wells

contamination_df = pd.DataFrame(0, index=range(height), columns=range(width))
contamination_df.rename(index={0:'Cells',1:'Platelets',2:'RBCs',3:'% cell with < 10% mito'}, inplace=True)
contamination_df.columns = np.array(sample_list['sample_names'])

for adata_index in range(len(RNA_adata)):
    adata = RNA_adata[adata_index]
    well = RNA_files[adata_index]
    print(well)
    #Remove platelets
    original_cell_no = adata.shape[0]
    
    print('Annotating platelet contamination and Megakaryocytes.')
    mat = csr_matrix(adata.X)
    mat = mat[:, adata.var_names.isin(['Pf4'])].todense()

    platelets = adata[np.ravel(mat > 0)].obs.index
    no_platelets = len(platelets)
    
    platelet_info = [False] * original_cell_no
    
    index = 0
    
    for x in adata.obs.index:
        if x in platelets:
            platelet_info[index] = True
        index += 1
        
    adata.obs['platelet'] = platelet_info   
    print('Annotating RBCs')

    mat = csr_matrix(adata.X)
    mat = mat[:, adata.var_names.isin(['Hbb'])].todense()
    
    RBCs = adata[np.ravel(mat > 1)].obs.index
    no_RBCs = len(RBCs)
    
    RBC_info = [False] * original_cell_no
    
    index = 0
    for x in adata.obs.index:
        if x in RBCs:
            RBC_info[index] = True
        index += 1

    adata.obs['RBC'] = RBC_info
    
    #Calculate quality metrics
    mito_genes = adata.var_names.str.startswith('mt-')
    adata.obs['percent_mito'] = np.sum(
        adata[:, mito_genes].X, axis=1).A1 / np.sum(adata.X, axis=1).A1
    #sc.pp.calculate_qc_metrics(adata, expr_type='counts', var_type='genes', percent_top=(50, 100, 200, 500), inplace=True)
    # for each cell compute fraction of counts in mito genes vs. all genes
    # the `.A1` is only necessary as X is sparse (to transform to a dense array after summing)

    # add the total counts per cell as observations-annotation to adata
    adata.obs['n_counts'] = adata.X.sum(axis=1).A1
    adata.obs['n_genes'] = (adata.X !=0).sum(axis=1).A1
    sc.pl.violin(adata, ['n_genes', 'n_counts', 'percent_mito'],
                 jitter=0.4, multi_panel=True)
    sc.pl.scatter(adata, x='n_counts', y='percent_mito')
    cells_mito_cutoff = np.sum([adata.obs['percent_mito'] < 0.1])/original_cell_no
    sc.pl.scatter(adata, x='n_counts', y='n_genes')
    
    contamination_df[np.array(sample_list['sample_names'])[adata_index]] = [original_cell_no, no_platelets, no_RBCs, cells_mito_cutoff]
    RNA_adata[adata_index] = adata
```
